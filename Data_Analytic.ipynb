```python
data.drop(['Name', 'Fare', 'PassengerId', 'Ticket', 'Cabin', 'Embarked'], inplace= True, axis = 1)
```

```python
data.rename(columns={'Number_of_siblings/Spouse':'siblings/Spouse'}, inplace=True)
```

```python
data.rename(columns={'Parch':'Parent_children'}, inplace=True)
```

```python
data['Survived'] = data['Survived'].replace(['No','Yes'],[0,1])
```

```python
data['Sex'] = data['Sex'].replace(['male', 'female'], [1,2])
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
data
```

```python
# In this code, the features are seperated from the label and given name X and y.
X = data.drop('Survived', axis = 1)
y = data['Survived']

# here, we create 4 new variables for both features and label and Train and test
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state= 30)

```

```python
X_train
```

```python
X_test
```

```python
# This is creating an obj of Logistic Regression
lr = LogisticRegression()
```

```python
# This code helps LR to learn from the train features and train label
lr.fit(X_train,y_train)
```

```python
computer_prediction = lr.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, computer_prediction)
```
